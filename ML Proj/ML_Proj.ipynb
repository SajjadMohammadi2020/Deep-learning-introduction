```python
## ایمپورت کردن کتابخانه های مورد نیاز برای برنامه
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
## ساخت دیتافریم از ایکس و ایگرگ
df = pd.DataFrame({'x' : [1.2 , 2.5 , 3.6] , 'y' : [8.5 , 9.2 , 10.1]})

## تبدیل ورودی و خروجی به آرایه نامپای
x = np.asanyarray(df[['x']])
y = np.asanyarray(df[['y']])
```

```python
a = b = 0
n = len(x)
sum_y  = sum(y)
sum_x2 = 0
sum_x = sum(x)
sum_xy = 0

## به دست آوردن پارامتر های رگرسیون بر مبنای فرمول به دست آوردن دستی پارامتر ها
for i in range(n) : 
    sum_xy += (x[i] * y[i])
    sum_x2 += (x[i]**2) 
b = ((sum_y * sum_x2)-(sum_x * sum_xy))/((n*sum_x2)-(sum_x**2))
a = ((n*sum_xy)-(sum_x*sum_y))/((n*sum_x2)-(sum_x**2))
print("y = ax + b")
print("a = " , a[0], "b = " , b[0])
```

```python
y_hat = []
q_err = 0
ae = 0

## محاسبه مقدار پیش بینی شده + خطای کوادراتیک + خطای قدرمطلق
for i in range(n) : 
    res = (a * x[i]) + b 
    y_hat.append(res)
    q_err += ((res-y[i])**2)
    ae += abs(res- y[i])
print("quadratic error : " , q_err[0])
print("absolute error : " , ae[0])
```

```python
## ترسیم نمودار برای دیتای اولیه و مقدار پیش بینی شده
plt.plot(x,y ,label="input" )
plt.plot(x,y_hat , label = "predict")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```
